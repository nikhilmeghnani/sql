```
spark.sql("""Create Table Sample_1
(
X Bigint,
Y Bigint,
Z Bigint);
""")
```

```
spark.sql("""Insert into Sample_1 Values (200,400,1);""")
spark.sql("""Insert into Sample_1 Values (200,400,2);""")
spark.sql("""Insert into Sample_1 Values (200,400,3);""")
spark.sql("""Insert into Sample_1 Values (10000,60000,1);""")
spark.sql("""Insert into Sample_1 Values (500,600,1);""")
spark.sql("""Insert into Sample_1 Values (500,600,2);""")
spark.sql("""Insert into Sample_1 Values (20000,80000,1);""")
```

```
spark.sql("""SELECT * from sample_1""").show()
```

```
spark.sql("""
WITH dup_records AS
(
select x, y, z, count(*) over (partition by x, y order by x) as dup_count from sample_1
)
select x, y, z from dup_records where dup_count > 1
""").show()
```
